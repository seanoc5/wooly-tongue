```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
# Initialize tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
# Example sentence
sent1 = "After stealing money from the bank, he went to the river bank to think."
sent2

# Tokenize the input sentence and create PyTorch tensors
inputs = tokenizer(sent1, return_tensors="pt")
outputs = model(**inputs)
```

```python
# Get the hidden states from the last layer
last_hidden_states = outputs.last_hidden_state

# Access embeddings for the specific word 'bank', considering its context
# Find all indices for 'bank' in the tokenized sentence
bank_tokens = [i for i, token in enumerate(inputs['input_ids'][0]) if tokenizer.decode([token]) == 'bank']
bank_tokens
```

```python
# Extract the embeddings for 'bank' at each occurrence
bank_embeddings = [last_hidden_states[0][index] for index in bank_tokens]

# Output embeddings for 'bank' (Note: each tensor represents 'bank' in a different context)
for i, emb in enumerate(bank_embeddings):
    print(f"Embedding for 'bank' occurrence {i+1}: {emb}")
```

This script:

    Tokenizes the input sentence.
    Feeds the tokenized input to BERT.
    Extracts the last layer of hidden states which represent the context-sensitive embeddings.
    Specifically extracts embeddings for occurrences of the word "bank," demonstrating how its representation changes depending on whether it's referring to a financial institution or the side of a river.

The resulting embeddings reflect the contextual meaning of the word within the sentence. This approach is especially useful in natural language understanding tasks where context plays a crucial role.
