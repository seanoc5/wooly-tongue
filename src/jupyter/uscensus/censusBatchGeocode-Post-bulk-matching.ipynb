```python
import os
import requests
import pandas as pd
import io
import csv
import glob
```

```python
path = '/opt/data/GIS-Census/florida'
# files = glob.glob(f"{path}/district_responses-*.csv")
files = glob.glob(f"{path}/data_result_*.csv")
```

```python
df_geocoded = []
for f in files:
    print(f"File: {f}")
    csv = pd.read_csv(f, header=None, on_bad_lines='skip')
    df_geocoded.append(csv)
df_geocoded = pd.concat(df_geocoded)
```

```python
df_geocoded.columns = ['VoterID', 'ADDRESS_IN', 'MATCH_INDICATOR', 'MATCH_TYPE', 'ADDRESS_OUT', 'LONG_LAT', 'TIGER_EDGE', 'STREET_SIDE', 'FIPS_STATE', 'FIPS_COUNTY', 'CENSUS_TRACT', 'CENSUS_BLOCK']
df_geocoded.shape
```

```python
df_geocoded.head()
```

```python
# Unique ID, Street address, City, State, ZIP
# soe_file = f"{path}/SoE-full.csv"
# soe_file = f"{path}/John_OConeco_original_less_history.csv"
soe_file = f"{path}/census-post-matching2.csv"
df_soe = pd.read_csv(soe_file, low_memory=False)
```

```python
df_soe.head()
```

```python
df_joined = df_soe.merge(df_geocoded, left_on='VoterID', right_on='VoterID', how='left')
```

```python
df_joined.head()
```

```python
outfile = f"{path}/census-post-matching3.csv"
df_joined.to_csv(outfile, index=False)
```

```python
df_unmatched = df_joined[df_joined['MATCH_INDICATOR'] != 'Match']
```
